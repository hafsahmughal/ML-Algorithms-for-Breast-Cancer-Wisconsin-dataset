### 6. Dimensionality Reduction with PCA

#### PCA for Modeling (Retaining 95% Variance)

This PCA will reduce the dimensionality while preserving 95% of the total variance in the dataset. The resulting components can be used as input for machine learning models.

```python
from sklearn.decomposition import PCA

# PCA to retain 95% of the variance
pca_95 = PCA(n_components=0.95)

# Ensure X_scaled_df has no NaN values before PCA
# This step addresses the ValueError: Input X contains NaN from previous runs
X_scaled_df_clean = X_scaled_df.dropna()

# If dropping rows, we also need to align the target variable 'y'
# Only keep the 'y' values corresponding to the rows that were NOT dropped.
# This assumes that if there are NaNs, they are row-wise.
# Check if any rows were dropped
if len(X_scaled_df_clean) < len(X_scaled_df):
    print(f"Warning: Dropped {len(X_scaled_df) - len(X_scaled_df_clean)} rows containing NaN values before PCA.")
    # Realign y if rows were dropped
    # The index of X_scaled_df and y should be aligned if created from the same df
    y_aligned = y.loc[X_scaled_df_clean.index]
else:
    y_aligned = y # No rows dropped, y remains as is

X_pca_95 = pca_95.fit_transform(X_scaled_df_clean)

print(f'Number of components retaining 95% variance: {pca_95.n_components_}')
print(f'Explained variance ratio by components: {pca_95.explained_variance_ratio_.sum():.2f}')
```

#### PCA for Visualization (2 Components)

This PCA will reduce the dataset to 2 principal components, which is ideal for visualizing the data in a 2D plot and observing potential clusters or patterns.

```python
pca_2 = PCA(n_components=2)
X_pca_2 = pca_2.fit_transform(X_scaled_df)

# Create a DataFrame for the 2 principal components for easier handling and visualization
X_pca_2_df = pd.DataFrame(data=X_pca_2, columns=['principal_component_1', 'principal_component_2'])

print('First 5 rows of the 2-component PCA DataFrame:')
display(X_pca_2_df.head())

print(f'Explained variance ratio by 2 components: {pca_2.explained_variance_ratio_.sum():.2f}')
```

### Plot: PCA with Eigenvectors

This plot visualizes the 2-dimensional PCA-transformed data points, colored by their `Diagnosis`. Overlaid on this plot are arrows representing the principal components (eigenvectors). The length and direction of these arrows indicate how much each original feature contributes to the variance captured by the principal components.

```python
import numpy as np
from matplotlib.patches import Ellipse

# Function to draw a confidence ellipse
def plot_ellipse(ax, x, y, ellipse_color, label, alpha=1.0, n_std=3.0):
    # Calculate covariance matrix
    cov = np.cov(x, y)
    # Calculate eigenvalues and eigenvectors
    lambda_, v = np.linalg.eig(cov)
    # Get angle of rotation for the ellipse
    theta = np.degrees(np.arctan2(*v[:, 0][::-1]))
    # Calculate width and height of ellipse (scaled by n_std)
    width, height = 2 * n_std * np.sqrt(lambda_)

    # Create ellipse patch with matching color outline and no fill
    ellipse = Ellipse(xy=(np.mean(x), np.mean(y)),
                      width=width, height=height,
                      angle=theta, edgecolor=ellipse_color, label=f'{label} Ellipse', alpha=alpha, fill=False, linewidth=2)
    ax.add_patch(ellipse)


plt.figure(figsize=(12, 10))

# Define a custom palette for clear differentiation
custom_palette = {'Benign': '#90EE90', 'Malignant': '#800080'} # Light green and purple

ax = sns.scatterplot(
    x='principal_component_1',
    y='principal_component_2',
    hue=y.map({0: 'Benign', 1: 'Malignant'}), # Map numerical diagnosis back to original labels
    data=X_pca_2_df,
    palette=custom_palette, # Use the custom palette
    alpha=1, # Make points solid
    ax=plt.gca()
)

# Get the feature names
feature_names = X.columns

# Get the components (eigenvectors) from the pca_2 object
components = pca_2.components_ # Shape (n_components, n_features)

# Calculate the absolute loadings for each feature on each component
abs_loadings = np.abs(components)

# Identify the single top feature for each component
# For PC1: Get index of the top feature based on absolute loading
top_pc1_index = np.argsort(abs_loadings[0])[::-1][:1]
# For PC2: Get index of the top feature based on absolute loading
top_pc2_index = np.argsort(abs_loadings[1])[::-1][:1]

# Combine these two unique indices
top_feature_indices = np.unique(np.concatenate((top_pc1_index, top_pc2_index)))

# Scale the eigenvectors for better visualization.
# The scaling factor can be adjusted based on the data spread.
# Here, we scale them by the maximum value of principal components to keep arrows within the plot boundaries.
# A scaling factor of 3 to 5 often works well.
scale_factor = 4 # Adjusted for visual clarity

for i in top_feature_indices:
    comp_x, comp_y = components.T[i]
    plt.arrow(0, 0, comp_x * scale_factor, comp_y * scale_factor, color='orange', width=0.03, head_width=0.1, head_length=0.1)
    # plt.text(comp_x * scale_factor * 1.1, comp_y * scale_factor * 1.1, feature_names[i], color='red', ha='center', va='center', fontsize=9)


# Add ellipses for each category
# Filter data for Benign (0) and Malignant (1)
benign_data = X_pca_2_df[y == 0]
malignant_data = X_pca_2_df[y == 1]

# Plot ellipses using the same colors from the custom_palette
plot_ellipse(ax, benign_data['principal_component_1'], benign_data['principal_component_2'], custom_palette['Benign'], 'Benign')
plot_ellipse(ax, malignant_data['principal_component_1'], malignant_data['principal_component_2'], custom_palette['Malignant'], 'Malignant')


# Manually create legend elements for clarity and consistency
legend_elements = [
    plt.Line2D([0], [0], marker='o', color='w', label='Benign', markerfacecolor=custom_palette['Benign'], markersize=10),
    plt.Line2D([0], [0], marker='o', color='w', label='Malignant', markerfacecolor=custom_palette['Malignant'], markersize=10),
    plt.Line2D([0], [0], color='orange', linestyle='-', linewidth=2, label='Eigenvector') # Add legend entry for eigenvector
]
ax.legend(handles=legend_elements, loc='best')

plt.title('2-Component PCA with Top Two Eigenvectors and Category Ellipses')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.savefig('pca_eigenvectors.png', bbox_inches='tight') # Save the plot
plt.show()
```

### 7. Splitting the Data into Training and Testing Sets

To evaluate the performance of our machine learning models, we need to split our dataset into training and testing sets. The training set will be used to train the model, and the testing set will be used to assess its generalization ability on unseen data. We will use `train_test_split` from `sklearn.model_selection` to achieve this, with an 80/20 split and a `random_state` for reproducibility.

```python
from sklearn.model_selection import train_test_split

# Split the PCA-transformed data (X_pca_95) and the target (y) into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca_95, y, test_size=0.2, random_state=42)

print(f'Shape of X_train: {X_train.shape}')
print(f'Shape of X_test: {X_test.shape}')
print(f'Shape of y_train: {y_train.shape}')
print(f'Shape of y_test: {y_test.shape}')
```

### 8. Model Building and Training

Now, we will train five different machine learning classification models on our preprocessed and dimensionality-reduced training data. For simplicity, we will use the default parameters for each model.

#### 8.1. Logistic Regression

Logistic Regression is a linear model used for binary classification. It models the probability of a binary outcome.

```python
from sklearn.linear_model import LogisticRegression

# Initialize and train the Logistic Regression model
log_reg_model = LogisticRegression(random_state=42, solver='liblinear') # Using 'liblinear' solver for better convergence with default parameters
log_reg_model.fit(X_train, y_train)

print('Logistic Regression Model Trained Successfully!')
```

#### 8.2. K-Nearest Neighbors (KNN)

KNN is a non-parametric, lazy learning algorithm. It classifies a new data point based on the majority class among its 'k' nearest neighbors in the training data.

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize and train the K-Nearest Neighbors model
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)

print('K-Nearest Neighbors Model Trained Successfully!')
```

#### 8.3. Decision Tree Classifier

A Decision Tree builds a classification model in the form of a tree structure. It breaks down a dataset into smaller and smaller subsets while at the same time an associated decision tree is incrementally developed.

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize and train the Decision Tree model
decision_tree_model = DecisionTreeClassifier(random_state=42)
decision_tree_model.fit(X_train, y_train)

print('Decision Tree Model Trained Successfully!')
```

#### 8.4. Random Forest Classifier

Random Forest is an ensemble learning method for classification (and regression) that operates by constructing a multitude of decision trees at training time and outputting the class that is the mode of the classes (classification) or mean prediction (regression) of the individual trees.

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and train the Random Forest model
random_forest_model = RandomForestClassifier(random_state=42)
random_forest_model.fit(X_train, y_train)

print('Random Forest Model Trained Successfully!')
```

#### 8.5. Support Vector Machine (SVM)

SVM is a powerful and versatile machine learning model capable of performing linear or non-linear classification, regression, and even outlier detection. It works by finding the optimal hyperplane that best separates classes in the feature space.

```python
from sklearn.svm import SVC

# Initialize and train the Support Vector Machine model
svm_model = SVC(random_state=42)
svm_model.fit(X_train, y_train)

print('Support Vector Machine Model Trained Successfully!')
```

### 9. Model Evaluation

Now that our models are trained, it's crucial to evaluate their performance on the unseen test data (`X_test`, `y_test`). We will use several common classification metrics to assess how well each model predicts the `Diagnosis`.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Dictionary to store evaluation metrics for each model
metrics = {}

models = {
    'Logistic Regression': log_reg_model,
    'K-Nearest Neighbors': knn_model,
    'Decision Tree': decision_tree_model,
    'Random Forest': random_forest_model,
    'Support Vector Machine': svm_model
}

for name, model in models.items():
    y_pred = model.predict(X_test)

    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    metrics[name] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'Confusion Matrix': cm
    }

    print(f'--- {name} Evaluation ---')
    print(f'Accuracy: {accuracy:.4f}')
    print(f'Precision: {precision:.4f}')
    print(f'Recall: {recall:.4f}')
    print(f'F1-Score: {f1:.4f}')
    print(f'Confusion Matrix:\n{cm}')

    # Plot Confusion Matrix
    plt.figure(figsize=(4, 3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
    plt.title(f'Confusion Matrix for {name}')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()
    print('\n')

# Create a DataFrame for easy comparison of metrics
metrics_df = pd.DataFrame({name: {k: v for k, v in data.items() if k != 'Confusion Matrix'} for name, data in metrics.items()}).T
print('--- Model Comparison ---')
display(metrics_df.sort_values(by='F1-Score', ascending=False))
```

### 9. Model Evaluation

Now that our models are trained, it's crucial to evaluate their performance on the unseen test data (`X_test`, `y_test`). We will use several common classification metrics to assess how well each model predicts the `Diagnosis`.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import time # Import the time module

# Dictionary to store evaluation metrics for each model
metrics = {}

models = {
    'Logistic Regression': log_reg_model,
    'K-Nearest Neighbors': knn_model,
    'Decision Tree': decision_tree_model,
    'Random Forest': random_forest_model,
    'Support Vector Machine': svm_model
}

# Re-train models to capture training time (or use already trained models and just record prediction time)
# For consistency, let's re-train them here while measuring the time.
# Note: The models were already trained, so this will re-run the training step.

for name, model in models.items():
    start_time = time.time()
    model.fit(X_train, y_train) # Re-fit the model to measure training time
    end_time = time.time()
    training_time = end_time - start_time

    y_pred = model.predict(X_test)

    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)

    metrics[name] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'Confusion Matrix': cm,
        'Training Time (s)': training_time # Add training time
    }

    print(f'--- {name} Evaluation ---')
    print(f'Accuracy: {accuracy:.4f}')
    print(f'Precision: {precision:.4f}')
    print(f'Recall: {recall:.4f}')
    print(f'F1-Score: {f1:.4f}')
    print(f'Training Time: {training_time:.4f}s') # Print training time
    print(f'Confusion Matrix:\n{cm}')

    # Plot Confusion Matrix
    plt.figure(figsize=(4, 3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
    plt.title(f'Confusion Matrix for {name}')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()
    print('\n')

# Create a DataFrame for easy comparison of metrics
metrics_df = pd.DataFrame({name: {k: v for k, v in data.items() if k != 'Confusion Matrix'} for name, data in metrics.items()}).T
print('--- Model Comparison ---')
display(metrics_df.sort_values(by='F1-Score', ascending=False))
```
